```python
correlation_matrix = df.select_dtypes(include=["number"]).corr()
correlation_matrix
plt.figure(figsize=(10,6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.show()
```

```python
# correlazioni importanti:

# 4046 e Total Volume positiva   # ovvio perchè all'aumentare degli avocado con quel codice venduti, aumenta il numero totale di avocado venduti (1)

# 4225 e Total Volume positiva # ovvio vedi (1)
# 4225 e 4046 positiva  # non sono in causa effetto, dovrei investigare altri valori per capire perchè sono in correlazione positiva. (2) 

# 4770 e Total Volume  positiva   # ovvio vedi (1)
# 4770 e 4046 positiva # vedi (2)
# 4770 e 4225 positiva # vedi (2)

# Total Bags e Total Volume  positiva # ovvio all' aumentare degli avocado venduti, aumentano il numero di bags totali (3)
# Total Bags e 4046 positiva # vedi (3) 
# Total Bags e 4225 positiva # vedi (3)
# Total Bags e 4770 positiva # vedi (3)

# Small Bags e  Total Volume positiva # all'aumentare degli avocado venduti(che siano tutti o con un codice specifico) aumentano il numero di buste(in generale, small, large e xlarge) (4) 
# Small Bags e 4046 positiva # vedi (4)
# Small Bags e 4225 positiva # vedi (4)
# Small Bags e 4770 positiva # vedi (4)
# Small Bags e Total Bags positiva # ovvio perchè le small bags sono contenute nelle total bags, quindi se aumentano le prime lo fanno anche le seconde. (5) 

# Large Bags e  Total Volume positiva
# Large Bags e 4046 positiva # vedi (4)
# Large Bags e 4225 positiva # vedi (4)
# Large Bags e 4770 positiva # vedi (4)
# Large Bags e Total Bags positiva # analogo a (5) ma per large bags
# Large Bags e Small Bags positiva # vedi (2)

# XLarge Bags e  Total Volume positiva
# XLarge Bags e 4046 positiva # vedi (4)
# XLarge Bags e 4225 positiva # vedi (4)
# XLarge Bags e 4770 positiva # vedi (4)
# XLarge Bags e Total Bags positiva   # analogo a (5) ma per xlarge bags
# XLarge Bags e Small Bags positiva # vedi (2)
# XLarge Bags e Large Bags positiva # vedi (2)



# correlazioni deboli:

# 4046 e AveragePrice leggermente negativa (-0.21) # all'aumentare dei 4046 venduti, diminuisce il prezzo medio leggermente,
# i 4046 sono i più piccoli, dunque immagino anche quelli che costano meno.
```

```python
# Nel 2017 sono stati raggiunti dei prezzi medi per tutti e 3 i tipi di avocado molto alti, 
#anche se il prezzo massimo è stato raggiunto nel 2016

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

plt.figure(figsize=(8, 5))
sns.countplot(x='year', data=df, hue='AveragePrice', palette='coolwarm')
plt.xlabel('year')
plt.ylabel('AveragePrice')
plt.title('')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()

# dal grafico è interessante notare che ci sia una ciclicità dei prezzi, che evidentemente segue il ciclo di maturazione del frutto.
# Infatti ogni anno si parte con il prezzo medio più basso e man mano aumenta il prezzo (all'allontanarsi dalla stagione di maturazione).
```

```python
# curiosità sulle quantità totali vendute per codice.
conta4046 = df['4046'].sum()
conta4225 =df['4225'].sum()
conta4770 =df['4770'].sum()

plt.figure(figsize=(7, 7))
plt.pie([conta4046, conta4225, conta4770], labels=['4046', '4225', '4770'], autopct='%1.1f%%',
        colors=['red', 'green', 'blue'], startangle=90, explode=[0 , 0, 0])
plt.title('')
plt.show()
```
